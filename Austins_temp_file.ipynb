Make Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Import whole data into DataFrame

```python
df_data = pd.read_csv('data/countries_non_pop.csv')
df_data
```

Begin cleaning the data: Starting with getting rid of fields not wanted

```python
df_canada = df_data[df_data['Country Name'].str.contains("Canada")]
df_japan = df_data[df_data['Country Name'].str.contains("Japan")]
df_japan
```

Set all given years to a variable

```python
years = ['1960', '1961', '1962', '1963', '1964', '1965', '1966', '1967', '1968',
       '1969', '1970', '1971', '1972', '1973', '1974', '1975', '1976', '1977',
       '1978', '1979', '1980', '1981', '1982', '1983', '1984', '1985', '1986',
       '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995',
       '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004',
       '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013',
       '2014', '2015']
```

Set new df for the birth rate and death rate indexed by year and combine them

```python
birth_rate_canada = df_canada[df_canada['Indicator Name'] == 'Birth rate, crude (per 1,000 people)'].iloc[:, 5:].T
death_rate_canada = df_canada[df_canada['Indicator Name'] == 'Death rate, crude (per 1,000 people)'].iloc[:, 5:].T
birth_rate_canada.columns = ['Birth Rate']
death_rate_canada.columns = ['Death Rate']
birth_rate_canada.index.name = 'Year'
death_rate_canada.index.name = 'Year'
combined_data_canada = birth_rate_canada.join(death_rate_canada)
combined_data_canada.index = combined_data_canada.index
```

Filter the data for every five years and plot it

```python
filtered_data = combined_data_canada[combined_data_canada.index.astype(int) % 5 == 0]
plt.figure(figsize=(20, 8))
plt.plot(filtered_data['Birth Rate'], marker='o', label='Birth Rate')
plt.plot(filtered_data['Death Rate'], marker='o', label='Death Rate')
plt.title('Canada Birth and Death Rates Over Time')
plt.xlabel('Year')
plt.ylabel('Rate (per 1,000 people)')
plt.legend()
plt.show()
```

Do the same for Cyprus

```python
birth_rate_japan = df_japan[df_japan['Indicator Name'] == 'Birth rate, crude (per 1,000 people)'].iloc[:, 5:].T
death_rate_japan = df_japan[df_japan['Indicator Name'] == 'Death rate, crude (per 1,000 people)'].iloc[:, 5:].T
birth_rate_japan.columns = ['Birth Rate']
death_rate_japan.columns = ['Death Rate']
birth_rate_japan.index.name = 'Year'
death_rate_japan.index.name = 'Year'
combined_data_japan = birth_rate_japan.join(death_rate_japan)
combined_data_japan.index = combined_data_japan.index
```
